```python
import matplotlib.pyplot as plt
import matplotlib.colors as pltc
from random import sample

all_colors = [k for k,v in pltc.cnames.items()]
fracs = np.random.randint(100, size=30)

for column in data.columns:
  if data[column].dtype == object:
    colors = sample(all_colors, len(fracs))
    print("For Column ", column)
    fig, ax = plt.subplots()
    data[column].value_counts().plot(ax=ax, kind='bar', xlabel=column, ylabel='frequency', color = colors, edgecolor='purple')
    plt.show()
    print(data[column].value_counts())
    print("=======================")
    print('\n ')
```

```python
%matplotlib inline 
import seaborn as sns 
import matplotlib.pyplot as plt 
fueltype_count = data['fueltype'].value_counts() 
sns.set(style="darkgrid") 
sns.barplot(fueltype_count.index, fueltype_count.values, alpha=0.9) 
plt.title('Frequency Distribution of fueltype type') 
plt.ylabel('Number of Occurrences', fontsize=12) 
plt.xlabel('fueltype', fontsize=12) 
plt.show() 
```

```python
Q1 = (data[[ 'symboling', 'CarName', 'fueltype',
            'aspiration', 'doornumber', 'carbody', 'drivewheel', 'enginelocation',
            'wheelbase', 'carlength', 'carwidth', 'carheight', 'curbweight', 'enginetype',
            'cylindernumber', 'enginesize', 'fuelsystem', 'boreratio', 'compressionratio',
            'horsepower', 'peakrpm', 'citympg', 'highwaympg', 'price']]).quantile(0.25)

Q3 = (data[[ 'symboling', 'CarName', 'fueltype',
            'aspiration', 'doornumber', 'carbody', 'drivewheel', 'enginelocation',
            'wheelbase', 'carlength', 'carwidth', 'carheight', 'curbweight', 'enginetype',
            'cylindernumber', 'enginesize', 'fuelsystem', 'boreratio', 'compressionratio',
            'horsepower', 'peakrpm', 'citympg', 'highwaympg', 'price']]).quantile(0.75)
            
IQR = Q3-Q1
maximum =  Q3+1.5*IQR

print('nilai maximum dari masing-masing variabel adalah : ')
print(maximum)

minimum = Q1-1.5*IQR
print('nilai minimum dari masing-masing variabel adalah : ')
print(minimum)

more_than = (data > maximum)
lower_than = (data < minimum)
data = data.mask(more_than, maximum, axis=1)
data = data.mask(lower_than, minimum, axis=1)
print('persebaran data setelah ditangani outlier : ')
print(data[['symboling', 'CarName', 'fueltype',
            'aspiration', 'doornumber', 'carbody', 'drivewheel', 'enginelocation',
            'wheelbase', 'carlength', 'carwidth', 'carheight', 'curbweight', 'enginetype',
            'cylindernumber', 'enginesize', 'fuelsystem', 'boreratio', 'compressionratio',
            'horsepower', 'peakrpm', 'citympg', 'highwaympg', 'price']].describe())
```

```python
q1 = data["price"].quantile(0.25)
q3 = data["price"].quantile(0.75)

iqr = q3-q1 #Interquartile range
fence_low  = q1-1.5*iqr
fence_high = q3+1.5*iqr

data.loc[(data["price"] < fence_low) | (data["price"] > fence_high)]
```

```python
q1 = data["wheelbase"].quantile(0.25)
q3 = data["wheelbase"].quantile(0.75)

iqr = q3-q1 #Interquartile range
fence_low  = q1-1.5*iqr
fence_high = q3+1.5*iqr

data.loc[(data["wheelbase"] < fence_low) | (data["wheelbase"] > fence_high)]
```
